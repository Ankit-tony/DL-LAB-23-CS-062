Comparision between list and tensors.

```python
import time
n=10000

a=list(range(n))
b=list(range(n))

t1=torch.arange(n)
t2=torch.arange(n)

start =time.time()
for i in range(n):
  a[i]*b[i]
end = time.time()
print("Using List : ",end-start)

start=time.time()
t1*t2
end = time.time()
print("Using tensors : ",end-start)
```

1D , 2D and 3D array using numpy.

```python
n_1d = np.array([1,3,4])
n_2d = np.array([[11,12,13],[4,5,6]])
n_3d = np.array([[[1,2,3],[4,5,6]],[[12,12,23],[45,56,67]]])

print(n_1d)
print(" ")
print(n_2d)
print(" ")
print(n_3d)
```

1D , 2D and 3D tensors using Pytorch.

```python
t_1d = torch.tensor([1,3,4])
t_2d = torch.tensor([[11,12,13],[4,5,6]])
t_3d = torch.tensor([[[1,2,3],[4,5,6]],[[12,12,23],[45,56,67]]])

print(t_1d)
print(" ")
print(t_2d)
print(" ")
print(t_3d)
```

Element wise operation using tensors.

```python
a = torch.tensor([1,2,3])
b = torch.tensor([4,5,6])

print("Addition : ",a+b)
print("Subtraction : ",a-b)
print("Multiplication : ",a*b)
print("Division : ",a/b)
```

Element wise operation using numpy array.

```python
a_np = np.array([1,2,3])
b_np = np.array([4,5,6])
print("Addition : ",a_np+b_np)
print("Subtraction : ",a_np-b_np)
print("Multiplication : ",a_np*b_np)
print("Division : ",a_np/b_np)
```

Indexing , Slicing operations in tensors.

```python
print("Indexing")
x = torch.tensor([[11,12,13],[14,15,16]])
print(x[0])
print(x[:,2])
print(x[1,2])
```

```python
print("Extracting Sub-tensors")
print(x[0:2,1:3])

print("Boolean Masking")
y = torch.tensor([1,2,3,4,5])

mask = y>2
print(mask)
print(y[mask])
y_np = np.array([1,2,3,4,5])
print(y[y>2])
```

.view , .reshape , .unsqueeze and squeeze function in pytorch.

```python
#.view()
t = torch.arange(12)
print(t)

t2 = t.view(3, 4)
print(t2)

#.reshape()
t3 = t.reshape(4, 3)
print(t3)

#.unsqueeze()
a = torch.tensor([1, 2, 3])
print(a.shape)

b = a.unsqueeze(0)
print(b.shape)

c = a.unsqueeze(1)
print(c.shape)

#.squeeze()
d = torch.tensor([[1, 2, 3]])
print(d.shape)

print(d.squeeze().shape)
```

.reshape in numpy

```python
np_arr = np.arange(12)
np_reshaped = np_arr.reshape(3, 4)
print(np_reshaped)
```

Broadcasting in tensors

```python
print("pytorch")
x = torch.tensor([[1, 2, 3],
                  [4, 5, 6]])

y = torch.tensor([10, 20, 30])

print(x + y)

print("numpy")
x_np = np.array([[1, 2, 3],
                 [4, 5, 6]])

y_np = np.array([10, 20, 30])

print(x_np + y_np)
```

In place vs out of place operations

```python
#out of place operations
a = torch.tensor([1, 2, 3])
b = a + 5

print(a)
print(b)

#in of place operations
a = torch.tensor([1, 2, 3])
a.add_(5)

print(a)
```
